```python
from scipy.stats import chi2_contingency
import pandas as pd

# Define the significance level (alpha). If p-value < alpha, we reject the null hypothesis.
alpha = 0.05

# List of all categorical features we want to test for association with charges.
# This list has been corrected to use the actual one-hot encoded BMI category columns.
cat_features = [
    'is_female',                      # Binary: 0 for male, 1 for female
    'is_smoker',                      # Binary: 0 for non-smoker, 1 for smoker
    'region_northwest',               # Binary: 1 if region is northwest, 0 otherwise
    'region_southeast',               # Binary: 1 if region is southeast, 0 otherwise
    'region_southwest',               # Binary: 1 if region is southwest, 0 otherwise
    'bmi_category_Obesity Class I',   # Binary: 1 if BMI is in Obesity Class I, 0 otherwise
    'bmi_category_Obesity Class II',  # Binary: 1 if BMI is in Obesity Class II, 0 otherwise
    'bmi_category_Obesity Class III', # Binary: 1 if BMI is in Obesity Class III, 0 otherwise
    'bmi_category_Overweight',        # Binary: 1 if BMI is in Overweight category, 0 otherwise
    'bmi_category_Underweight'        # Binary: 1 if BMI is in Underweight category, 0 otherwise
]

# Bin the numerical 'charges' column into 4 equal-sized quantiles.
# This converts 'charges' into an ordinal categorical variable (0, 1, 2, 3).
# This step is necessary because Chi-squared tests require categorical data.
df_cleaned['charges_bin'] = pd.qcut(df_cleaned['charges'], q=4, labels=False)

# Initialize an empty dictionary to store the results of the Chi-squared tests.
chi2_results = {}

# Loop through each categorical feature to perform the Chi-squared test.
for col in cat_features:
    # Create a contingency table (cross-tabulation) between the current categorical feature
    # and the 'charges_bin' column. This table shows the frequency distribution of the two variables.
    contingency = pd.crosstab(df_cleaned[col], df_cleaned['charges_bin'])

    # Perform the Chi-squared test using chi2_contingency.
    # It returns:
    # chi2_stat: The Chi-squared test statistic.
    # p_val: The p-value of the test.
    # _: Degrees of freedom.
    # _: Expected frequencies.
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)

    # Determine whether to reject or accept the null hypothesis.
    # The null hypothesis (H0) states that there is no association between the two categorical variables.
    # If p_val < alpha, we reject H0, meaning there is a statistically significant association.
    # If p_val >= alpha, we accept H0, meaning there is no statistically significant association.
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'

    # Store the results for the current feature in the chi2_results dictionary.
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

# Convert the results dictionary into a Pandas DataFrame for better readability and analysis.
chi2_df = pd.DataFrame(chi2_results).T

# Sort the DataFrame by p-value in ascending order.
# Features with lower p-values (closer to 0) have stronger evidence against the null hypothesis
# and thus a stronger association with the 'charges_bin' variable.
chi2_df = chi2_df.sort_values(by='p_value')

# Display the resulting DataFrame.
display(chi2_df)
```
